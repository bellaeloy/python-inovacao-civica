### Leia o arquivo dos dados da Receita Federal e cruze eles com os dados da CEAP

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
# 1. LENDO OS DADOS DA RECEITA FEDERAL
```

```python
url_rf = "https://serenata-de-amor-data.nyc3.digitaloceanspaces.com/2019-11-19-companies.csv.xz"
```

```python
#alterar o tipo do dado do cnpj para ele ser lido como str
DTYPE = {
    "cnpj":str
}
```

```python
df_rf = pd.read_csv(url_rf, compression = "xz", dtype = DTYPE)
```

```python
df_rf.head() #Visualizar o df
```

```python
# 2. LER OS DADPS DA CEAP
```

```python
url_ceap = "http://www.camara.leg.br/cotas/Ano-2019.csv.zip"
```

```python
DTYPE = {
    'txNomeParlamentar': str,
    'ideCadastro': str,
    'nuCarteiraParlamentar': str,
    'nuLegislatura': str,
    'sgUF': str,
    'sgPartido': str,
    'codLegislatura': str,
    'numSubCota': str,
    'txtDescricao': str,
    'numEspecificacaoSubCota': str,
    'txtDescricaoEspecificacao': str,
    'txtFornecedor': str,
    'txtCNPJCPF': str,
    'txtNumero': str,
    'indTipoDocumento': str,
    'datEmissao': str,
    'vlrDocumento': float,
    'vlrGlosa': str,
    'vlrLiquido': float,
    'numMes': str,
    'numAno': str,
    'numParcela': str,
    'txtPassageiro': str,
    'txtTrecho': str,
    'numLote': str,
    'numRessarcimento': str,
    'nuDeputadoId': str,
    'ideDocumento': str,
}
```

```python
df_ceap = pd.read_csv(url_ceap, sep=";", compression = "zip", dtype = DTYPE)
```

```python
df_ceap.head()
```

```python
#Antes de dar o merge tem que garantir que as colunas correspondem, para isso, usar o replace.
```

```python
#'\D' é uma expressão regular de todos os caracteres não numericos no regex
```

```python
#regex pode ser usado para procurar especificidades.
```

```python
df_ceap["txtCNPJCPF"] = df_ceap["txtCNPJCPF"].str.replace(r"\D", "", regex=True) #sobescrever
```

```python
#Remover os cpfs NULOS
```

```python
filtro_Nan = df_ceap.ideCadastro.isna()
df_ceap[~filtro_Nan].shape
```

```python
df_ceap.loc[filtro_Nan, "ideCadastro"] = "-1"
```

```python
df_ceap.loc[filtro_Nan, "ideCadastro"]
```

```python
filtro_Nan_Partido = df_ceap.sgPartido.isna()
df_ceap.loc[filtro_Nan_Partido , "sgPartido"] = "-1"
```

```python
df_ceap.sgPartido
```

```python
#ALTERAR A DATA DE EMISSÃO e sobrescrever
```

```python
df_ceap["datEmissao"] = pd.to_datetime(df_ceap.datEmissao, format="%Y-%m-%d")
```

```python
#Somar CEAP com o da receita federal
```

```python
df_gastos_finais = df_ceap.merge(df_rf, how = "left", left_on = "txtCNPJCPF", right_on = "cnpj")
```

```python
#Remover os duplicados
```

```python
df_gastos_finais.shape
```

```python
df_gastos_finais.drop_duplicates().shape
```

### Some o total da coluna `vlrLiquido` agrupando por `txtCNPJCPF` e `txtFornecedor`

```python
df_gastos_finais.groupby(["txtCNPJCPF", "txtFornecedor"])["vlrLiquido"].sum().to_frame()
```

### Ordene os dados da consulta anterior de forma decrescente

```python
df_gastos_finais.groupby(["txtCNPJCPF", "txtFornecedor"])["vlrLiquido"].sum().sort_values(ascending =False).to_frame()
```

### Consulte se houveram gastos reembolsados associados a empresas que na época estavam com status `08`


status == 8 significa `BAIXADA` (*a empresa foi extinta, seja por solicitação dos sócios ou por ação da Receita*)

```python
df_gastos_finais.situation.unique()
```

```python
#converter as datas para poder comparar
```

```python
df_gastos_finais["situation_date"] = pd.to_datetime(df_gastos_finais.situation_date, format = "%Y-%m-%d")
```

```python
#filtrar as empresas baixadas
```

```python
filtro_baixadas = df_gastos_finais.situation == 8
df_gastos_finais[filtro_baixadas].shape
```

```python
#Analise para ver se alguma empresa baixada foi usada
```

```python
df_gastos_finais[filtro_baixadas].query("datEmissao > situation_date")
```

```python
#verificar quais parlamentares fizeram isso
```

```python
df_gastos_finais[filtro_baixadas].query("datEmissao > situation_date").groupby("txNomeParlamentar").agg({"ideDocumento": "count", "vlrLiquido":"sum"}).sort_values(by ="vlrLiquido", ascending = False)
```
