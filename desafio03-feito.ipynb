### 1) Acesse a api da camara, e recupere os reembolsos do ano de 2018

```python
import requests

# fazendo o download do conteúdo do arquivo
url = "http://www.camara.leg.br/cotas/Ano-2018.csv.zip"
r = requests.get(url)
```

```python
# abrir o arquivo
file = open("Ano-2018.csv.zip","wb")
file.write(r.content)
file.close()
```

```python
#descompactar o arquivo
from zipfile import ZipFile
zip_file = ZipFile("Ano-2018.csv.zip", 'r') #criar o objeto que representa o zip usando o modo de leitura r
zip_file.extract(member = "Ano-2018.csv") #pode adicionar o 'path = nomeDaPasta"
zip_file.close()
```
